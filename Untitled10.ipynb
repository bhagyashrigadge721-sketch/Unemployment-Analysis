```python
# -------------------------------------------------------------------------
# STEP 1: Import Required Data Science Libraries
# -------------------------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Data Analytics Libraries Loaded Successfully ---")

# -------------------------------------------------------------------------
# STEP 2: Load the Unemployment Dataset
# -------------------------------------------------------------------------
# Loading the primary dataset file uploaded in your Colab sidebar
df = pd.read_csv('Unemployment in India.csv')

print("\n--- Initial 5 Records of the Dataset ---")
display(df.head())

# -------------------------------------------------------------------------
# STEP 3: Data Cleaning & Formatting
# -------------------------------------------------------------------------
# Stripping hidden blank spaces from column headers
df.columns = df.columns.str.strip()

# Drop rows containing all missing null fields safely
df.dropna(inplace=True)

# Convert the Date feature column into a proper Python Date-Time object
df['Date'] = pd.to_datetime(df['Date'].str.strip(), dayfirst=True)

# Extract Months and Years for tracking seasonal patterns
df['Month'] = df['Date'].dt.strftime('%B')
df['Year'] = df['Date'].dt.year

# -------------------------------------------------------------------------
# STEP 4: Analyze Covid-19 Impact & Trends (Data Visualization)
# -------------------------------------------------------------------------
print("\n--- Generating Statistical Unemployment Visualizations ---")
plt.figure(figsize=(15, 6))

# Line Plot: Timeline trend of Estimated Unemployment Rate (%) across 2019-2020
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df, hue='Area', marker='o', linewidth=2)

# Marking the Covid-19 Lockdown phase to satisfy point number 3 of your task rules
plt.axvspan('2020-03-24', '2020-06-30', color='red', alpha=0.15, label='Covid-19 National Lockdown Peak')
plt.title('Timeline Trend of Unemployment Rate in India (Pre-Covid vs Covid-19 Peak)')
plt.xlabel('Timeline Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

# -------------------------------------------------------------------------
# STEP 5: Seasonal Trends & Regional Area Distribution
# -------------------------------------------------------------------------
plt.figure(figsize=(14, 5))

# Plot 1: Comparing Rural vs Urban structural distribution
plt.subplot(1, 2, 1)
sns.boxplot(x='Area', y='Estimated Unemployment Rate (%)', data=df, palette='Pastel1')
plt.title('Unemployment Spread: Rural vs Urban Sectors')

# Plot 2: Average monthly seasonal patterns across India
plt.subplot(1, 2, 2)
sns.barplot(x='Month', y='Estimated Unemployment Rate (%)', data=df, errorbar=None, palette='muted')
plt.xticks(rotation=45)
plt.title('Average Monthly Seasonal Unemployment Trends')

plt.tight_layout()
plt.show()

# -------------------------------------------------------------------------
# STEP 6: Policy Insights Summary Generation
# -------------------------------------------------------------------------
print("\n--- Key Policy & Economic Insights ---")
state_unemp = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).head(5)
print("Top 5 Hardest Hit Regions/States (Average %):")
print(state_unemp)
```
